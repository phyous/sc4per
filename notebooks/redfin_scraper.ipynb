```
import http.client

conn = http.client.HTTPSConnection("realtor.p.rapidapi.com")

headers = {
    'x-rapidapi-host': "realtor.p.rapidapi.com",
    'x-rapidapi-key': "<ENTER>"
    }

conn.request("GET", "/properties/v2/list-sold?sort=sold_date&city=Seattle&offset=0&state_code=WA&limit=10000", headers=headers)

res = conn.getresponse()
data = res.read()
s = data.decode("utf-8")
```

```
import json

dict = json.loads(s)
```

```
data = {}
for l in dict['properties']:  
  try:
    data[l['property_id']] = l
  except:
    print(f"Error parsing: {l}")
```

```
import json 
with open('test.json', 'w') as f:
  f.write(json.dumps(data))

!cat test.json
```
